```python
class Social_bot:
    def __init__(self):
        # instance/class variables
        # Facebook page url
        self.login_page = "https://www.facebook.com/"

        # Facebook Page Menu Url
        self.page_ref = "https://www.facebook.com/pages/?category=your_pages&ref=bookmarks"

        

        # By default session is None, will set after Starting the chrome
        self.browser_session = None

        # By default is set to 0, if any website visits then will be increment by one, only set to integer
        self.browser_visit = 0

        # By default is set to 0, which represents no login :
        self.login = 0

        # Xpath For Facebook login id - email
        self.user_xpath = '//*[@id="email"]'

        # Xpath for Facebook login password
        self.pass_xpath = '//*[@id="pass"]'
        
        # For facebook login id credentials
        self.user = None

        # For Facebook login password credentials
        self.password = None

        # Text posting content variable Will be set via text_posting_content_load function, By default set to None
        self.textContents = None


    def initiate_chrome(self):
    # This function is to start chrome. Successful login will return 1

        if self.browser_session is None:
            self.browser_session = webdriver.Firefox()
            return 1
        else:
            return -1


    def close_session(self):
    # This function is to close the session. Successful operation will return 1

        if self.browser_session is None:
            return -1
        else:
            try:
                self.browser_session.close()
                self.browser_session.quit()
                self.browser_session = None
                return 1
            except Exception as e:
                print(f"Error closing session: {str(e)}")
                return -1


    def page_load(self, path=None):
    # This function is to load page,successful operation will return None

        if path is None:
            if self.browser_session is None:
                raise ValueError("Browser session not started.")
            else:
                try:
                    self.browser_session.get(self.login_page)
                    self.browser_visit += 1
                except Exception as e:
                    print(f"Error loading page: {str(e)}")
                    raise
        else:
            if self.browser_session is None:
                raise ValueError("Browser session not started.")
            else:
                try:
                    self.browser_session.get(path)
                    self.browser_visit += 1
                except Exception as e:
                    print(f"Error loading page: {str(e)}")
                    raise


    def do_login(self):
        # if self.browser not started, or already logged in, then it will return -1
        if self.browser_session is None or self.login != 0:
            return -1

        # Load credentials from JSON file
        try:
            with open("credentials_load.json") as filePointer:
                contents = json.load(filePointer)

            self.user = contents["Email Address"]
            self.password = contents["Password"]

            # Use self.browser_session for finding elements
            self.browser_session.find_element(By.XPATH, self.user_xpath).send_keys(self.user)
            self.browser_session.find_element(By.XPATH, self.pass_xpath).send_keys(self.password, Keys.ENTER)
            self.browser_visit += 1
            self.login = 1
            print("Login Successful ...........")
            return 1
        except Exception as e:
            print(f"Error during login: {str(e)}")
            return -1


    def page_navigation_partial(self, pg_name):

        if self.browser_session is None:
            return -1
        else:
            try:
                self.browser_session.find_element(By.XPATH,pg_name).click()
                self.browser_visit += 1
                return 1
            except Exception as e:
                print(f"Error during page navigation: {str(e)}")
                return -1


    def credential_loads_using_json(self):

        # This function is to load credentials from json file
        # Will help to set credentials instead of manual entry 
        try:
            with open("credentials_load.json") as filePointer:
                contents = filePointer.read()

            contents = json.loads(contents)

            
            self.user = contents["Email Address"]
            self.password = contents["Password"]
            
            # Print the loaded data
            print("Loaded data from JSON file:")
            
            print(f"Email Address: {self.user}")
            print(f"Password: {self.password}")

            # for freeing the json memory part
            del(contents)
            return 1
        except:
            return -1



    def list_all_groups(self):
        # Wait for the content to load
        WebDriverWait(self.browser_session, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'div.x9f619.x78zum5.x1r8uery.xdt5ytf.x1iyjqo2.xs83m0k.x150jy0e.x1e558r4.xjkvuk6.x1iorvi4.xnpuxes')))

        # Extract article links and titles
        group_elements = self.browser_session.find_elements(By.CSS_SELECTOR,'a.x1i10hfl.xjbqb8w.x6umtig.x1b1mbwd.xaqea5y.xav7gou.x9f619.x1ypdohk.xt0psk2.xe8uvvx.xdj266r.x11i5rnm.xat24cr.x1mh8g0r.xexx8yu.x4uap5.x18d9i69.xkhd6sd.x16tdsg8.x1hl2dhg.xggy1nq.x1a2a7pz.x1heor9g.xt0b8zv.x1s688f')
        group_name = [element.text for element in group_elements]
        group_link = [element.get_attribute('href') for element in group_elements]
        groups_df = pd.DataFrame(columns = ["Group Name", "Group Link"])
        groups_df["Group Name"] = group_name
        groups_df["Group Link"] = group_link

        print("Retrieved all Groups in Facebook to Database................")
        return groups_df
        
    
    def group_post(self, dataframe, message_txt, photo_path_txt):
        
        # Check if the session is logged in
        if self.browser_session is None:
            return -1
        else: 
            for index, row in dataframe.iterrows():
                group_name = row['Group Name']
                group_link = row['Group Link']
                self.browser_session.get(group_link)
                # Add text to the post
                post_path = "/html/body/div[1]/div/div[1]/div/div[3]/div/div/div[1]/div[1]/div/div[2]/div/div/div[4]/div/div[2]/div/div/div[1]/div[1]/div/div/div/div[1]/div"
                # Wait for the element to be present and visible
                self.browser_session.find_element(By.XPATH,post_path).click()
                print("Creating Your Post Rn......")

                # # Scroll to the element using JavaScript
                # self.browser_session.execute_script("arguments[0].scrollIntoView(true);", element)


                post_box_xpath = "//*[contains(@class, 'notranslate') and contains(@class, '_5rpu')]"
                post_box = WebDriverWait(self.browser_session, 15).until(EC.presence_of_element_located((By.XPATH, post_box_xpath)))
                post_box.send_keys(message_txt)
                
                
                # # Add photo to the post
                # add_photo_btn_xpath = "/html/body/div[1]/div/div[1]/div/div[4]/div/div/div[1]/div/div[2]/div/div/div/div/div[1]/form/div/div[1]/div/div/div[1]/div/div[3]/div[1]/div[2]/div[1]/div/span/div/div/div[1]/div/div/div[2]"
                # add_photo_btn = WebDriverWait(self.browser_session, 15).until(EC.presence_of_element_located((By.XPATH, add_photo_btn_xpath))) 
                # add_photo_btn.click()     

                
                # add_photo_btn_xpath = "/html/body/div[1]/div/div[1]/div/div[4]/div/div/div[1]/div/div[2]/div/div/div/div/div[1]/form/div/div[1]/div/div/div[1]/div/div[3]/div[1]/div[2]/div[1]/div/span/div/div/div[1]/div/div/div[2]"
                # add_photo_btn = WebDriverWait(self.browser_session, 15).until(
                #     EC.presence_of_element_located((By.XPATH, add_photo_btn_xpath))
                # )
                # add_photo_btn.click().send_keys(photo_path_txt)

                # Post: 
                time.sleep(5)
                post_btn_xpath = "/html/body/div[1]/div/div[1]/div/div[4]/div/div/div[1]/div/div[2]/div/div/div/div/div[1]/form/div/div[1]/div/div/div[1]/div/div[3]/div[3]/div[1]/div/div/div[2]"
                post_btn = WebDriverWait(self.browser_session, 15).until(
                    EC.element_to_be_clickable((By.XPATH, post_btn_xpath))
                )
                post_btn.click()

                
                
                self.browser_session.back()
                time.sleep(5)
```
