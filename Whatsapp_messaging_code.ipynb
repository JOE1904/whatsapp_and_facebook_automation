```python
from urllib.parse import quote 
from selenium import webdriver 
from selenium.webdriver.common.action_chains import ActionChains 
from selenium.webdriver.common.keys import Keys
import time
```

## **Approach 1: For Messaging:**

```python
with open("message.txt",'r') as file: 
    msg = file.read()

msg = quote(msg)

numbers = []
with open('numbers.txt','r') as file: 
    for num in file.readlines():
        numbers.append(num.rstrip())

driver = webdriver.Firefox()

link = "https://web.whatsapp.com"

driver.get(link)
time.sleep(15)

for num in numbers: 
    link2 = f'https://web.whatsapp.com/send/?phone=91{num}&text={msg}'
    driver.get(link2)
    time.sleep(38)
    action = ActionChains(driver)
    action.send_keys(Keys.ENTER)
    action.perform()
    time.sleep(10)

driver.close()
driver.quit()
```

## **Code for Sending Whatsapp Messages Individually and based on Group IDs:**

```python
import pywhatkit as kit 
import pyautogui as pg
from datetime import datetime
```

```python
# Sending Message to Single Contact: 
def send_single_contacts(message_file_path,phone_numbers_file_path):
    if type(message_file_path) != str:
        print("Upload File Path as string........")
    else: 
        pass 
    
    with open(message_file_path,'r') as file: 
        msg = file.read()

    numbers = []
    with open(phone_numbers_file_path,'r') as file: 
        for num in file.readlines():
            numbers.append(num.rstrip())

    for num in numbers:
        # Get the current date and time
        current_time = datetime.now()

        # Extract the hour and minute
        hour = current_time.hour
        minute = current_time.minute
        try:
            kit.sendwhatmsg(f"+91{num}", message= msg,time_hour=hour, time_min=minute+1, wait_time=7)
            pg.press("enter")
            print(f"Message Sent on +91 - {num}")
            print()
        except Exception as e:
            print(e) 

# Sending Message to Groups:

def group_message(message_file_path,group_ids_file_path):
    
    if type(message_file_path) != str:
        print("Upload File Path as string........")
    else: 
        pass 
    
    with open(message_file_path,'r') as file: 
        msg = file.read()

    group_ids = []
    with open(group_ids_file_path,'r') as file: 
        for id in file.readlines():
            group_ids.append(id.rstrip())
            print(id)
    
    for id in group_ids:
        current_time = datetime.now()
        # Extract the hour and minute
        hour = current_time.hour
        minute = current_time.minute
        try:
            kit.sendwhatmsg_to_group(group_id=f"{id}",message= msg,time_hour=hour, time_min=minute+1, wait_time=7)
            print(f"Message sent to Group: {id}")
        except Exception as e:
            print(e) 
```

```python
group_message(group_ids_file_path="group_ids.txt",message_file_path="message.txt")
```
